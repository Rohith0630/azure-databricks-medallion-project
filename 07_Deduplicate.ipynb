```python
# 1. Read your streaming table from yesterday
silver_df = spark.read.table("f1_processed.silver_incremental_results")

# 2. The "Senior" Way: Deduplicate based on Unique Keys
# (One driver can only have ONE result per race)
clean_silver_df = silver_df.dropDuplicates(["raceId", "driverId"])

# 3. Save to a final 'Cleaned' table
clean_silver_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.f1_processed.silver_cleaned")

print(f"✅ Cleaned Table Created! Rows after deduplication: {clean_silver_df.count()}")
```
