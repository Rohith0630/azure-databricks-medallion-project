```python
# 1. Define the specific volume path from your screenshot
volume_path = "/Volumes/workspace/default/raw_data/raw_data/circuits.csv"

# 2. Read with Schema (EPAM Q27 - why Dataframe over RDD? Because of Schema!)
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

circuit_schema = StructType(fields=[
    StructField("circuitId", IntegerType(), False),
    StructField("name", StringType(), True),
    StructField("location", StringType(), True),
    StructField("country", StringType(), True)
])

# 3. Load into your 'f1_processed' schema shown in the sidebar
df = spark.read \
    .option("header", True) \
    .schema(circuit_schema) \
    .csv(volume_path)

df.write.format("delta").mode("overwrite").saveAsTable("f1_processed.silver_circuits")
```

```python
import time
from functools import wraps

# 🎁 The Decorator (The Wrapper)
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs) # Runs your actual code
        end = time.time()
        print(f"⏱️ Execution Time for {func.__name__}: {end - start:.2f}s")
        return result
    return wrapper

# 🏎️ Using it on your F1 Ingestion
@timer
def load_f1_data():
    # This represents your Spark code
    print("Reading CSV and writing to Delta...")
    time.sleep(2) # Simulating a 2-second job
    return "Success"

# Execute
load_f1_data()
```
