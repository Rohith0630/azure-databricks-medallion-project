```python
%sql
-- 1. Create the Schema inside your new Catalog
CREATE SCHEMA IF NOT EXISTS f1_dev.presentation;

-- 2. Move (Clone) your existing table into the new Enterprise structure
-- We use 'DEEP CLONE' to make a full copy
CREATE TABLE f1_dev.presentation.gold_race_results 
DEEP CLONE f1_presentation.gold_race_results;

-- 3. Verify it's there
SELECT * FROM f1_dev.presentation.gold_race_results LIMIT 5;
```

```python
%sql
-- Create a new table based on your gold results
CREATE TABLE f1_dev.presentation.top_drivers_lineage AS
SELECT driver_name, points
FROM f1_dev.presentation.gold_race_results
WHERE points > 50;
```
